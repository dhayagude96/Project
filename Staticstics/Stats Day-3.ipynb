```python
import numpy as np
import seaborn as sns
import pandas as pd


dataset = [1,1,1,2,2,3,3,3,5,5,6,7,8,9,9,10,11,30,25]

"""
n=

Q1 = (25/100) x (n+1) = 

Q3 = (75/100) x (n+1) =

IQR = Q3 - Q1 = 

LF = Q1 - 1.5 X IQR =

HF = Q3 + 1.5 x IQR = 

Min = 

Max = 

"""
```

```python
#To calulate 1st & 3rd Quartile

Q1 , Q3 = np.percentile(dataset,[25,75])

print("Q1 Value of Dataset  : " , Q1)
print("Q3 Value of Dataset : " , Q3)

IQR = Q3 - Q1

print("IQR of Dataset : " , IQR)

LF = Q1 - 1.5 * IQR
HF = Q3 + 1.5 * IQR

print("Lower Fence : " , LF)
print("Higher Fence : " , HF)
print("Min value : " , np.min(dataset))
#print("Min value : " , np.max(dataset))
```
