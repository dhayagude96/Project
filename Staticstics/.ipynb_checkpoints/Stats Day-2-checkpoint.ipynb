# Percentiles and quartiles

Percentiles and quartiles are statistical measures used to describe the distribution of data in a dataset, particularly in the field of descriptive statistics. They help us understand the relative position of a particular data point or a range of data points within a dataset. Here are the definitions of percentiles and quartiles

# Percentile:

A percentile is a statistical measure that indicates the relative rank or position of a specific data point within a dataset when the data is sorted in ascending order.
It represents the percentage of data points in the dataset that are below or equal to the given data point.
For example, the 75th percentile (often denoted as P75) is the value below which 75% of the data points in the dataset fall.

# Quartile:

Quartiles are a specific type of percentiles that divide a dataset into four equal parts, each containing 25% of the data points.

The three quartiles are typically denoted as Q1 (the 25th percentile), Q2 (the 50th percentile, also known as the median), and Q3 (the 75th percentile).

Q1 represents the value below which 25% of the data points fall.

Q2 (the median) represents the value below which 50% of the data points fall.

Q3 represents the value below which 75% of the data points fall.

The interquartile range (IQR) is often used in conjunction with quartiles. It is the difference between Q3 and Q1, and it provides a measure of the spread or dispersion of the middle 50% of the data in the dataset.

Five number summary 

Minimum: This is the smallest value in the dataset, representing the lower extreme of the data.

First Quartile (Q1): This is the value that separates the lowest 25% of the data when sorted in ascending order. It is the 25th percentile.

Median (Second Quartile, Q2): This is the middle value of the dataset when it is ordered. It is also the 50th percentile and represents the point below which 50% of the data falls.

Third Quartile (Q3): This is the value that separates the lowest 75% of the data when sorted in ascending order. It is the 75th percentile.

Maximum: This is the largest value in the dataset, representing the upper extreme of the data.

The five-number summary is particularly useful for describing the central tendency, spread, and skewness of a dataset, and it is often used in box plots (box-and-whisker plots) to visualize the distribution of data. In a box plot, the minimum and maximum values are represented by whiskers, Q1 and Q3 define the edges of the box, and the median is marked within the box. This visualization helps in identifying outliers and understanding the overall shape of the data distribution.







Lower fenec & Higher fence

Lower Fence:

The lower fence is a boundary below which data points are considered potential outliers.
To calculate the lower fence, you typically subtract 1.5 times the Interquartile Range (IQR) from the First Quartile (Q1)

Upper Fence:

The upper fence is a boundary above which data points are considered potential outliers.
To calculate the upper fence, you typically add 1.5 times the Interquartile Range (IQR) to the Third Quartile (Q3)
