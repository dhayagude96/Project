```python
import pandas as pd

data = pd.read_html("https://www.timeanddate.com/weather/india/pune/ext")

df = data[0]

df.to_csv("C:/Users/Vishal/Desktop/Data/pune_report.csv")

Data = pd.read_csv("C:/Users/Vishal/Desktop/Data/pune_report.csv")

#Use Row info as Column Name

Data.columns = Data.iloc[0] 

#Delete Row form Data base

Data.drop(Data.index[0] , inplace = True) #Single Data at time

Data.drop(Data.index[15] , inplace = True)#Single Data at time

df = Data.drop(Data.index[2:8]) # #Multiple Data at time


#Rename column

Data.rename(columns={"Unnamed: 1_level_1":"Not AVA" , "Unnamed: 6_level_1":"Temp"} , inplace = True)


#Delete column


Data.drop(['Not AVA'] , axis =1 , inplace = True)

Data.drop(['Temp' ,'Amount'] , axis =1 , inplace = True)

```
